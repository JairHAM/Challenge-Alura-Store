### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

ingresos = {
    'Tienda': nombres_tiendas,
    'Ingresos Totales': [t['Precio'].sum() for t in tiendas]
}
df_ingresos = pd.DataFrame(ingresos)

print("--- Ingresos Totales por Tienda ---")
print(df_ingresos.to_string(index=False))
```

# 2. Ventas por categoría

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

categorias = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Categoría Más Vendida': [t['Categoría del Producto'].mode()[0] for t in tiendas]
}
df_categorias = pd.DataFrame(categorias)

print("--- Categoría Más Vendida por Tienda ---")
print(df_categorias.to_string(index=False))
```

# 3. Calificación promedio de la tienda


```python
tiendas = [tienda, tienda2, tienda3, tienda4]

reviews = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Promedio de Reseñas': [t['Calificación'].mean() for t in tiendas]
}
df_reviews = pd.DataFrame(reviews)

print("--- Promedio de Reseñas por Tienda ---")
print(df_reviews.to_string(index=False))
```

# 4. Productos más y menos vendidos

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

unidades_vendidas = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Total de Unidades Vendidas': [t.shape[0] for t in tiendas]
}
df_unidades = pd.DataFrame(unidades_vendidas)

print("--- Total de Unidades Vendidas por Tienda ---")
print(df_unidades.to_string(index=False))
```

# 5. Envío promedio por tienda

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

envio_promedio = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo de Envío Promedio': [t['Costo de envío'].mean() for t in tiendas]
}
df_envio = pd.DataFrame(envio_promedio)

print("--- Costo de Envío Promedio por Tienda ---")
print(df_envio.to_string(index=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingresos Totales': [1150880400.0, 1116343500.0, 1098019600.0, 1038375700.0],
    'Promedio de Reseñas': [3.976685, 4.037304, 4.048326, 3.995759],
    'Costo de Envío Promedio': [26018.69, 25216.24, 24805.68, 23459.46],
    'Total de Unidades': [2359, 2359, 2359, 2358]
}
df_resumen = pd.DataFrame(data)

plt.style.use('ggplot')

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(df_resumen['Tienda'], df_resumen['Ingresos Totales'] / 1000000, color='skyblue', alpha=0.8)
ax1.set_xlabel('Tienda')
ax1.set_ylabel('Ingresos Totales (en millones de $)', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_title('Rendimiento de Ingresos y Satisfacción por Tienda', fontsize=16)

ax2 = ax1.twinx()
ax2.plot(df_resumen['Tienda'], df_resumen['Promedio de Reseñas'], color='red', marker='o', linewidth=3)
ax2.set_ylabel('Promedio de Reseñas (1-5)', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.set_ylim(3.9, 4.1)

plt.show()

plt.figure(figsize=(8, 8))

df_resumen['% Contribución'] = df_resumen['Ingresos Totales'] / df_resumen['Ingresos Totales'].sum()

colores = ['#4CAF50', '#2196F3', '#FFC107', '#E91E63']
explotar = [0, 0, 0, 0.15]

plt.pie(df_resumen['% Contribución'],
        labels=df_resumen['Tienda'],
        autopct='%1.1f%%',
        startangle=90,
        colors=colores,
        explode=explotar,
        wedgeprops={'edgecolor': 'black'})

plt.title('Contribución Porcentual de Ingresos a la Cadena', fontsize=16)
plt.show()


plt.figure(figsize=(10, 6))

tamanio_punto = df_resumen['Ingresos Totales'] / 2500000

plt.scatter(df_resumen['Promedio de Reseñas'],
            df_resumen['Costo de Envío Promedio'],
            s=tamanio_punto,
            c=df_resumen['Costo de Envío Promedio'],
            cmap='viridis_r',
            alpha=0.8,
            edgecolors='black')

for i, txt in enumerate(df_resumen['Tienda']):
    plt.annotate(txt,
                 (df_resumen['Promedio de Reseñas'][i], df_resumen['Costo de Envío Promedio'][i]),
                 textcoords="offset points", xytext=(5,5), ha='center', fontsize=12, fontweight='bold')

plt.title('Eficiencia Operativa: Costo de Envío vs. Satisfacción (Tamaño = Ingreso)', fontsize=16)
plt.xlabel('Promedio de Reseñas (Peor Satisfacción a la Izquierda)')
plt.ylabel('Costo de Envío Promedio (Mayor Gasto Arriba)')

plt.grid(True, linestyle='--', alpha=0.6)
plt.colorbar(label='Costo de Envío Promedio')
plt.show()
```

```python
import matplotlib.pyplot as plt

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores = ['blue', 'green', 'red', 'orange']
plt.style.use('ggplot')

print("--- Gráfico 1: Distribución Geográfica de Ventas ---")

plt.figure(figsize=(12, 10))
plt.title('Distribución Geográfica de Ventas por Tienda (Tamaño = Precio del Producto)', fontsize=16)

for i, t in enumerate(tiendas):
    tamanio_punto = t['Precio'] / 500000

    plt.scatter(t['lon'],
                t['lat'],
                label=nombres_tiendas[i],
                c=colores[i],
                s=tamanio_punto,
                alpha=0.6,
                edgecolors='black',
                linewidths=0.5)

plt.xlabel('Longitud (lon)')
plt.ylabel('Latitud (lat)')
plt.legend(scatterpoints=1, frameon=True, title='Tiendas')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()


print("\n--- Gráfico 2: Distribución Geográfica de Reseñas (Tienda 1) ---")

plt.figure(figsize=(12, 8))
plt.scatter(tienda['lon'],
            tienda['lat'],
            c=tienda['Calificación'],
            cmap='RdYlGn',
            alpha=0.8,
            edgecolors='black')

plt.colorbar(label='Calificación (1=Malo, 5=Excelente)')
plt.title('Distribución Geográfica de las Reseñas (Tienda 1)', fontsize=16)
plt.xlabel('Longitud (lon)')
plt.ylabel('Latitud (lat)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
